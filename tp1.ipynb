```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
ecommerce_customers = pd.read_csv("ecommerce_customers.csv")
ecommerce_order_items = pd.read_csv("ecommerce_order_items.csv")
ecommerce_orders = pd.read_csv("ecommerce_orders.csv")
ecommerce_products = pd.read_csv("ecommerce_products.csv")
ecommerce_reviews = pd.read_csv("ecommerce_reviews.csv")
ecommerce_website_logs = pd.read_csv("ecommerce_website_logs.csv")
```

```python
def data_profiling(df):
    print (f'\n {df.head()}')
    print('___________________________________________________________________________')
    print (f'Shape : {df.shape} ')
    print(f"Memory : {df.memory_usage(deep=True).sum() / (1024**2):.2f} MB")
    
    print('___________________________________________________________________________')
    print(' missing values \n')
    print (f' {df.isnull().sum()}')
    print('___________________________________________________________________________')
    print('data types \n')
    print (df.dtypes)
    print('___________________________________________________________________________')
    print ('unique values')
    print(f'  {df.nunique()}')
    print('___________________________________________________________________________') 
    print(f'Duplicated rows : {df.duplicated().sum()}')    
    for col in df.columns:
        dups = df[col].duplicated().sum()
        print(f"{col}: {dups} duplicates")


data_profiling(ecommerce_customers)   
    
```
